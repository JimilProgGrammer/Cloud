```python
import pandas as pd
m = pd.read_csv('3DIMG_07NOV2019_0000_L1C_SGP.tif.csv')
n = pd.read_csv('3DIMG_07NOV2019_0030_L1C_SGP.tif.csv')
```

```python
print(m.head())
print(n.head())
```

```python
from math import sqrt
def eucledian(a1, a2):
    a1.pop()
    a2.pop()
    sumx=0
    for i in range(0, len(a1)):
        sumx+=((a1[i]-a2[i])*(a1[i]-a2[i]))
    return sqrt(sumx)

from scipy import spatial
x=m.to_dict('records')
y=n.to_dict('records')

for input1 in x:
    prev=9999
    solve=0
    next_x=0
    next_y=0
    for input2 in y:
        #result = 1 - spatial.distance.cosine(list(input1.values()), list(input2.values()))
        result=eucledian(list(input1.values()), list(input2.values()))
        if result < prev:
            solve=input2['label']
            next_x=input2['com_x']
            next_y=input2['com_y']
        prev=min(prev, result)
    print(input1['label'], prev, solve)
    pred_x=(input1['com_x']+next_x)/2
    pred_y=(input1['com_y']+next_y)/2
    print(pred_x, pred_y)
```

```python
def mpa(t1, t2, t):
    pred=[]
    pred.append((t1[0] + t2[0]) / 2.0)
    pred.append((t1[1] + t2[1]) / 2.0)
    
    pred_next=[]
    pred_next.append((t[0] + pred[0]) / 2.0)
    pred_next.append((t[1] + pred[1]) / 2.0)
    
    pred_next[0] = pred_next[0] + t[0] - pred[0]
    pred_next[1] = pred_next[1] + t[1] - pred[1]
    
    return pred_next
mpa([52, 352], [77, 299], [30, 156])
```

```python
from math import sqrt
def speed(t1, t2):
    diff_x=abs(t1[0]-t2[0])
    diff_y=abs(t1[1]-t2[1])
    dist=sqrt(diff_x*diff_x + diff_y*diff_y)*2.58*1000
    time=30
    print(dist/time)
speed([600,800], [595, 799])
```

```python
def directionx(t1, t2):
    if t1[1]-t2[1] > 0:
        direction="south "
    else:
        direction="north "
        
    if t1[0]-t2[0] > 0:
        direction+="west"
    else:
        direction+="east"
        
    
    return direction
directionx([600,800], [595, 799])
```
